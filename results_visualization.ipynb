```python
import numpy as np
import matplotlib.pyplot as plt


def moving_average(arr, window_size):
    return np.convolve(arr, np.ones(window_size)/window_size, mode='valid')

# Load the .npy file
reward_MAPPO = np.load("data_train\MAPPO_env_UAV network_number_1_consider_cloud_True.npy")
# reward_MAPPO = reward_MAPPO[0:3000]


mov_avr_reward_MAPPO = moving_average(reward_MAPPO, window_size=30)
# eps = np.linspace(0,3000,len(mov_avr_reward_MAPPO))

plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(8,6))
plt.plot(reward_MAPPO, color = 'red', linewidth=3, label= 'MAPPO')
plt.grid()
plt.xlabel('Training episode',fontsize = 18)
plt.ylabel('Average Episode Return',fontsize = 18)
plt.legend()
plt.tight_layout()
plt.savefig('episode_return.pdf')
```

```python
percentage_users_UAV0 = np.load('percentage_users_UAV0.npy')  # Total number of users = 400
percentage_users_UAV1 = np.load('percentage_users_UAV1.npy')  # Total number of users = 400
percentage_users_UAV2 = np.load('percentage_users_UAV2.npy')  # Total number of users = 400
percentage_users_UAV0_no_allocation = np.load('percentage_users_UAV0 copy.npy') # Total number of users = 400
percentage_users_UAV1_no_allocation = np.load('percentage_users_UAV1 copy.npy') # Total number of users = 400
percentage_users_UAV2_no_allocation = np.load('percentage_users_UAV2 copy.npy') # Total number of users = 400

plt.figure(figsize=(8,6))
plt.plot(percentage_users_UAV0, color = 'blue', linewidth=2, label='With IRS Allocation UAV0')
plt.plot(percentage_users_UAV1, color = 'green', linewidth=2, label='With IRS Allocation UAV1')
plt.plot(percentage_users_UAV2, color = 'orange', linewidth=2, label='With IRS Allocation UAV2')
# plt.plot(percentage_users_UAV0_no_allocation, color = 'red', linewidth=2, label='Without IRS Allocation UAV0',linestyle='--')
# plt.plot(percentage_users_UAV1_no_allocation, color = 'purple', linewidth=2, label='Without IRS Allocation UAV1',linestyle='--')
# plt.plot(percentage_users_UAV2_no_allocation, color = 'brown', linewidth=2, label='Without IRS Allocation UAV2',linestyle='--')
plt.grid()
plt.xlabel('Time slot (s)',fontsize = 18)
plt.ylabel('The number of users served',fontsize = 18)

plt.legend()
plt.tight_layout()
plt.savefig('percentage_users.png')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(percentage_users_UAV0_no_allocation, color = 'red', linewidth=2, label='Without IRS Allocation UAV0',linestyle='--')
plt.plot(percentage_users_UAV1_no_allocation, color = 'purple', linewidth=2, label='Without IRS Allocation UAV1',linestyle='--')
plt.plot(percentage_users_UAV2_no_allocation, color = 'brown', linewidth=2, label='Without IRS Allocation UAV2',linestyle='--')
plt.grid()
plt.xlabel('Time slot (s)',fontsize = 18)
plt.ylabel('The number of users served',fontsize = 18)

plt.legend()
plt.tight_layout()
plt.savefig('percentage_users.png')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(percentage_users_UAV0+percentage_users_UAV1+percentage_users_UAV2, color = 'blue', linewidth=4, label='With IRS Allocation')
plt.plot(percentage_users_UAV0_no_allocation+percentage_users_UAV1_no_allocation+percentage_users_UAV2_no_allocation, color = 'green', linewidth=4, label='Without IRS Allocation')
print("Average supported users with IRS allocation: ",np.mean(percentage_users_UAV0+percentage_users_UAV1+percentage_users_UAV2))
print("Average supported users without IRS allocation: ",np.mean(percentage_users_UAV0_no_allocation+percentage_users_UAV1_no_allocation+percentage_users_UAV2_no_allocation))
plt.grid()
plt.xlabel('Time slot (s)',fontsize = 18)
plt.ylabel('The number of users served',fontsize = 18)

plt.legend()
plt.tight_layout()
plt.savefig('percentage_users.png')
plt.show()
```
